```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.random.random_sample(25)
y = np.random.random_sample(25)
```

```python
sns.scatterplot(x=x, y=y);
plt.savefig('img/scatterplot.png')
```

```python
sns.regplot(x=x, y=y, ci=None);
plt.savefig('img/regression_line.png')
```
